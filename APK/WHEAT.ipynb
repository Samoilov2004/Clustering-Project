# Библиотеки и импорты

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("./dataset.csv")
print(df.shape)
df.head()
```

# EDA

Посмотрим вообще какое распределение имеют колонки эти вот 

```python
fig, axes = plt.subplots(4, 2, figsize=(10, 16))

# Плоский список осей для удобства
axes = axes.flatten()

# Построение графиков
for i, column in enumerate(df.columns):
    sns.histplot(df[column], kde=True, ax=axes[i])
    axes[i].set_title(f'Гистограмма для {column}')

# Настройка расположения графиков
plt.tight_layout()
plt.savefig('file4.png')
```

```python
correlation_matrix = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, linecolor='gray', cbar_kws={'shrink': 0.8})

plt.title('Correlation Matrix', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.savefig('file5.png')
plt.show()
```

# Препроцессинг

```python
def make_clean(list_of_columns, scaler_type = 1):
    df_preclean = df[list_of_columns]

    if scaler_type == 1:
        scaler = StandardScaler()
        df_scaled = scaler.fit_transform(df_preclean)

    df_result = pd.DataFrame(df_scaled, columns=list_of_columns)
    
    return df_result 
```

```python
# Проба оставлять разные столбцы
without_target = ['A', 'P', 'C', 'LK', 'WK', 'A_Coef', 'LKG']
experiment1 = ['A', 'C', 'A_Coef', 'LKG']
experiment2 = ['A', 'C', 'LK', 'A_Coef', 'LKG']

df_clean = make_clean(experiment1)
df_clean.head()
```

# Определение количества кластеров

```python
ssd = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=20)
    kmeans.fit(df_clean)
    ssd.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(K, ssd, marker='o', linestyle='-', color='b', label='SSD')
plt.xlabel('Количество кластеров')
plt.ylabel('SSD')
plt.title('Оптимальное количество клстеров методом локтя')
plt.legend()
plt.grid(True)

plt.savefig('file6.png')
plt.show()
```

# Построение моделей

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=20)
kmeans.fit(df_clean)
```

```python
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score, homogeneity_score, completeness_score, v_measure_score
```

```python
true_labels = df.target
predicted_labels = kmeans.predict(df_clean)

ari = adjusted_rand_score(true_labels, predicted_labels)
nmi = normalized_mutual_info_score(true_labels, predicted_labels)
homogeneity = homogeneity_score(true_labels, predicted_labels)
completeness = completeness_score(true_labels, predicted_labels)
v_measure = v_measure_score(true_labels, predicted_labels)

print(f"Adjusted Rand Index (ARI): {ari:.4f}")
print(f"Normalized Mutual Information (NMI): {nmi:.4f}")
print(f"Homogeneity: {homogeneity:.4f}")
print(f"Completeness: {completeness:.4f}")
print(f"V-Measure: {v_measure:.4f}")
```
